```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("../data/internet_usage.csv", parse_dates=["data"])
```

```python
# Visualizao bsica
plt.figure(figsize=(12, 5))
plt.plot(df['data'], df['consumo_mb'], label='Consumo Dirio (MB)')
plt.title("Consumo de Internet - 2024")
plt.xlabel("Data")
plt.ylabel("Consumo (MB)")
plt.legend()
plt.show()
```
